```python
#import packages
import pandas as pd
```

```python
#read in csv
bb = pd.read_csv(r"../data/Billboard1920.csv")
bb
```

```python
#strip /single
bb['song'] = bb['song'].str.replace('/single', '').str.strip('/').str.strip()
bb
```

```python
#split on /
bb[['track', 'artist']] = bb['song'].str.split('/', expand= True)
bb
```

```python
#drop 'song' column
bb = bb.drop(columns='song')
bb
```

```python
#replace weird strings with spaces. supposed to strip end spaces too and replace X with spaces, but that doesnt work

replace_dict = {'%20':' ', '%3A':' ', '%28':' ', '%29':' ', '%26':' ', '%21':' ', '%27':' ', '%24':'$', '%22':' ', '%40':' ', '%2C':' ', '%2B':' ', '%2F':' ', '%2A':' ', '%3F':' ', ' X ':' ', ' x ':' ', 'Mr.':' '}
bb['track'] = bb['track'].replace(replace_dict, regex= True).str.strip()
bb['artist'] = bb['artist'].replace(replace_dict, regex= True).str.strip()
bb
```

```python
#replace Xs separately, which works for some reason
bb['artist'] = bb['artist'].str.replace(' X ', ' ').str.replace(' x ', ' ')
```

```python
#lets try to strip end spaces separately
bb['track'] = bb['track'].str.strip()
bb['artist'] = bb['artist'].str.strip()
# bb.to_csv('testXreplace', index= False)
```

```python
#replace 'Featuring' with an empty string because the API throws an error if you don't
bb['artist'] = bb['artist'].str.replace('Featuring', '')
bb
```

```python
#drop duplicates
nodupes = bb.drop_duplicates(subset= ('track', 'artist'))
nodupes
```
