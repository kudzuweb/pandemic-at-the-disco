```python
#import packages
import pandas as pd
```

```python
#read in csv
bb = pd.read_csv(r"/Users/mauriaparker/Documents/NSS_Projects/Solo/pandemic-at-the-disco/Billboard1920.csv")
bb
```

```python
#strip /single
bb['song'] = bb['song'].str.replace('/single', '').str.strip('/').str.strip()
bb
```

```python
#split on /
bb[['track', 'artist']] = bb['song'].str.split('/', expand= True)
bb
```

```python
#drop 'song' column
bb = bb.drop(columns='song')
bb
```

```python
#replace weird strings with spaces

replace_dict = {'%20':' ', '%3A':' ', '%28':' ', '%29':' ', '%26':' ', '%21':' ', '%27':' ', '%24':'$', '%2C':' ', '%2B':' ', '%2F':' ', '%2A':' ', '%3F':' ', ' X ':' ', ' x ':' '}
bb['track'] = bb['track'].replace(replace_dict, regex= True).str.strip()
bb['artist'] = bb['artist'].replace(replace_dict, regex= True).str.strip()
bb
```

```python
#drop duplicates
nodupes = bb.drop_duplicates('track')
nodupes
```

```python
#replace 'Featuring' with an empty string because the API throws an error if you don't
nodupes['artist'] = nodupes['artist'].str.replace('Featuring', '')
nodupes.head(15)
```

```python
#export nodupes to csv
nodupes.to_csv('nodupesBillboard1920.csv', index= False)
```
