```python
#import packages
import pandas as pd
```

```python
#read in csv
cv = pd.read_csv(r"../data/Provisional_COVID-19_Death_Counts_by_Week_Ending_Date_and_State_20240104.csv")
```

```python
#drop extraneous columns
cv = cv.drop(columns=['Data as of', 'Start Date', 'End Date', 'MMWR Week', 'Group', 'State', 'Year', 'Month', 'Pneumonia Deaths', 'Influenza Deaths', 'Pneumonia and COVID-19 Deaths', 'Pneumonia, Influenza, or COVID-19 Deaths', 'Footnote'])
cv
```

```python
#drop state rows and US rows after 2020

cv = cv.drop(cv.index[52:])
cv
```

```python
#split out date column

cv[['Month', 'Day', 'Year']] = cv['Week Ending Date'].str.split('/', expand= True)
cv
```

```python
#drop date and day columns
cv = cv.drop(columns=['Week Ending Date', 'Day'])
cv
```

```python
#move year column to front via pop > insert
col = cv.pop('Year')
cv.insert(0, 'Year', col)
cv
```

```python
#move month column to front via pop > insert
col = cv.pop('Month')
cv.insert(1, 'Month', col)
cv
```

```python
#rename columns to labels that work for code
cv.rename(columns={'Year':'year', 'Month':'month', 'COVID-19 Deaths':'deaths', 'Total Deaths':'allcausedeaths', 'Percent of Expected Deaths':'percentexpected'})
```

```python
#export csv
cv.to_csv('../data/covidmaster.csv', index= False)
```
