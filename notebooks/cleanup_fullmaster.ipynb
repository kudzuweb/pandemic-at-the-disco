```python
import pandas as pd
```

```python
fm = pd.read_csv('../data/fullmaster.csv')
fm
```

```python
#split the date into year, month, and day
fm[['year', 'month', 'day']] = fm['date'].str.split('-', expand= True)
fm 
```

```python
#move year column to front via pop > insert
col = fm.pop('year')
fm.insert(0, 'year', col)
print(fm)
```

```python
#move month column to front via pop > insert
col = fm.pop('month')
fm.insert(1, 'month', col)
print(fm)
```

```python
#move day column to front via pop > insert
col = fm.pop('day')
fm.insert(2, 'day', col)
print(fm)
```

```python
#drop date column
fm = fm.drop(columns= 'date')
fm
```

```python
#export csv
fm.to_csv('../data/fullmasterfinal.csv', index= False)
```
