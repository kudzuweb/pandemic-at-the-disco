```python
import pandas as pd
```

```python
#import full Billboard csv
dupes = pd.read_csv(r"../data/cleanBillboard1920.csv")
dupes
```

```python
#import masterframe csv
mf = pd.read_csv(r"../data/masterframe.csv")
mf
```

```python
#many-to-one merge

fullmaster = pd.merge(dupes, mf[['track', 'artist', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'id', 'duration_ms', 'time_signature', 'error']], on=['track', 'artist'], how= 'left')
fullmaster
```

```python
#export fullmaster csv
fullmaster.to_csv('../data/fullmaster.csv', index= False)
```
